# Data Augmentation

```python
# data augmetation 
# https://github.com/mdbloice/Augmentor
import Augmentor
p = Augmentor.Pipeline("C:/Users/Jinxi/Desktop/EE211/Project/prune/unaugmented/N")
#p.ground_truth("./images/mask")

p.skew(probability=0.6,magnitude = 0.3)
p.flip_random(probability=0.5)
p.rotate(probability=0.5, max_left_rotation=5, max_right_rotation=5)
p.flip_left_right(probability=0.5)
p.zoom_random(probability=0.5, percentage_area=0.95)
p.flip_top_bottom(probability=0.5)

p.random_distortion(probability=1, grid_width=5, grid_height=5, magnitude=4)

p.sample(1000)

# name format
# _groundtruth_(1)_t2_a1b7301d-3930-4554-8d52-e52b023bee39.png
#         t2_original_a1b7301d-3930-4554-8d52-e52b023bee39
```
